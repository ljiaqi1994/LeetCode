```python
'''
给你一个用字符数组 tasks 表示的 CPU 需要执行的任务列表。其中每个字母表示一种不同种类的任务。任务可以以任意顺序执行，并且每个任务都可以在 1 个单位时间内执行完。在任何一个单位时间，CPU 可以完成一个任务，或者处于待命状态。

然而，两个 相同种类 的任务之间必须有长度为整数 n 的冷却时间，因此至少有连续 n 个单位时间内 CPU 在执行不同的任务，或者在待命状态。

你需要计算完成所有任务所需要的 最短时间 。

 

示例 1：

输入：tasks = ["A","A","A","B","B","B"], n = 2
输出：8
解释：A -> B -> (待命) -> A -> B -> (待命) -> A -> B
     在本示例中，两个相同类型任务之间必须间隔长度为 n = 2 的冷却时间，而执行一个任务只需要一个单位时间，所以中间出现了（待命）状态。 
示例 2：

输入：tasks = ["A","A","A","B","B","B"], n = 0
输出：6
解释：在这种情况下，任何大小为 6 的排列都可以满足要求，因为 n = 0
["A","A","A","B","B","B"]
["A","B","A","B","A","B"]
["B","B","B","A","A","A"]
...
诸如此类
示例 3：

输入：tasks = ["A","A","A","A","A","A","B","C","D","E","F","G"], n = 2
输出：16
解释：一种可能的解决方案是：
     A -> B -> C -> A -> D -> E -> A -> F -> G -> A -> (待命) -> (待命) -> A -> (待命) -> (待命) -> A
 

提示：

1 <= task.length <= 104
tasks[i] 是大写英文字母
n 的取值范围为 [0, 100]

来源：力扣（LeetCode）
链接：https://leetcode-cn.com/problems/task-scheduler
著作权归领扣网络所有。商业转载请联系官方授权，非商业转载请注明出处。
'''

class Solution:
    def leastInterval(self, tasks, n):
        total=0  #总时间
        total+=len(tasks)  #先加总现有任务所需时间
        tasks_count={}  #未排队的任务计数字典
        tasks_queue=[]  #任务栈
        for task in tasks:
            tasks_count[task]=tasks_count.get(task,0)+1  #对tasks作第一次的统计
        tasks_sorted = sorted(tasks_count.items(), key=lambda d: d[1], reverse=True)  #将任务计数字典按从高到低排序
        tasks_queue.append(tasks_sorted[0][0])  #第一次将任务排进栈
        tasks_count[tasks_sorted[0][0]]-=1 
        if tasks_count[tasks_sorted[0][0]]==0:
            tasks_count.pop(tasks_sorted[0][0])  #未排队的任务计数字典作相应调整（-1）
        tasks_sorted = sorted(tasks_count.items(), key=lambda d: d[1], reverse=True)  #重新将任务计数字典按从高到低排序
        while tasks_count:  #开始排栈
            FLAG=True  #指示是否要待机（排0）
            for j in range(len(tasks_sorted)):
                flag=True  #指示当前任务是否可排
                for k in range(n):
                    if k+1>len(tasks_queue):
                        break  #任务栈已经遍历完
                    elif tasks_queue[-k-1]==tasks_sorted[j][0]:
                        flag=False
                        break  #还不能做当前任务
                if flag:
                    tasks_queue.append(tasks_sorted[j][0])
                    tasks_count[tasks_sorted[j][0]]-=1
                    if tasks_count[tasks_sorted[j][0]]==0:
                        tasks_count.pop(tasks_sorted[j][0])
                    tasks_sorted = sorted(tasks_count.items(), key=lambda d: d[1], reverse=True)
                    FLAG=False
                    break
            if FLAG:
                tasks_queue.append("0")
                total+=1
        return total
```
